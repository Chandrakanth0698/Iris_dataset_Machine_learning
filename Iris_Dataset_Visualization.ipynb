```python
import numpy as np # for linear algebra
import pandas as pd # for data processing
#for data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read Data set
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
iris.info()
# we can there 6 colums different datatypes with 0 null values,
# so this make it easy as we dont need to handle missing or null values
```

```python
iris.describe()
#bascially we may not need 'id' column so its better to drop it
```

```python
iris=iris.drop('Id',axis=1)
iris.head()
```

```python
#we will check different types of species available to predict later in the project
iris['Species'].unique()
```

```python
#we create a scatter plot with speal length& width  features of data
iris.plot(kind='scatter', x='SepalLengthCm',y='SepalWidthCm')
```

```python
#joinplot is one the plot in seaborn which combines bivariate sctterplots and univerate histograms in the same figure
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris)
```

```python
iris.shape
```

```python
# giving different color to different species using seaborn facegrid
sns.FacetGrid(iris,hue='Species',height=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
#we can also directly call the hue parameter in sns.scatterplot without using facegrid
sns.scatterplot(data=iris,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
```

# lets create box plots for different featues and observe it with relative to Species

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=iris)
```

```python
sns.boxplot(x='Species',y='PetalWidthCm',data=iris)
```

```python
sns.boxplot(x='Species',y='SepalWidthCm',data=iris)
```

```python
sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
```

# lets see where most of the points fall on the boxplot with petal length since it is more distungished from one another

```python
#if we use jitter=True we can see all the points and they dont fall on 1 vertical line
#saving the resulting axis to ax each time overlaps the plots
ax=sns.boxplot(x='Species',y='PetalLengthCm',data=iris)
ax=sns.stripplot(x='Species',y='PetalLengthCm',data=iris,hue='Species',legend=False)
```

```python
# Vilion plot combines both the boxplot and strip plot and simplifies them
#denser regions of data are fatter
sns.violinplot(x='Species',y='PetalLengthCm',data=iris)
```

```python
#KDE Kernal Desity Estimator 
#kdeplot shows univariate relations of underlying featurs with kernal density estimate
sns.kdeplot(data=iris,x='PetalLengthCm',hue='Species',fill=True)
```
